```python
from lxml import etree as ET
from xml.etree.ElementTree import parse, Element
import codecs
import csv

k=[-1]
with open('test_labels_minmax.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    for row in readCSV:
        
        
        try:
            if k[-1]!=int(row[0]):
                f=codecs.open('image{}.xml'.format('{}'.format(7496+int(row[0]))),'w')
                
                root = ET.Element('annotation')
                filename = ET.SubElement(root, 'filename')
                filename.text='image{}'.format(str(7496+int(row[0])))
                path = ET.SubElement(root, 'path')
                path.text='C:\\Users\\Spock\\craters\\images\\image{}.jpg'.format(row[0])
                size = ET.SubElement(root, 'size')
                height = ET.SubElement(size, 'height')
                height.text = '224'
                width = ET.SubElement(size, 'width')
                width.text = '224'



                objet = ET.SubElement(root, 'object')
                name = ET.SubElement(objet, 'name')
                name.text = 'crater'

                bndbox = ET.SubElement(objet, 'bndbox')
                xmin = ET.SubElement(bndbox, 'xmin')
                xmin.text = '{}'.format(row[1])
                ymin = ET.SubElement(bndbox, 'ymin')
                ymin.text = '{}'.format(row[2])
                xmax = ET.SubElement(bndbox, 'xmax')
                xmax.text = '{}'.format(row[3])
                ymax = ET.SubElement(bndbox, 'ymax')
                ymax.text = '{}'.format(row[4])


 
                ET.tostring(root, pretty_print=True, xml_declaration=True)

                f.write(ET.tostring(root, pretty_print=True, xml_declaration=True).decode("utf-8") )
                f.close
                k.append(int(row[0]))
            else:
                doc = parse('image{}.xml'.format('{}'.format(7496+int(row[0]))))
                root = doc.getroot()
                e = Element('object')
                #e.text = 'This is a test'
                root.insert(2, e)
                
                name1 = Element('name')
                name1.text = 'crater'
                e.insert(1, name1)
                
                bnd1 = Element('bndbox')
                #name1.text = 'crater'
                e.insert(1, bnd1)
                
                xmin1 = Element('xmin')
                xmin1.text = '{}'.format(row[1])
                bnd1.insert(1, xmin1)
                
                ymin1 = Element('ymin')
                ymin1.text = '{}'.format(row[2])
                bnd1.insert(1, ymin1)
                
                xmax1 = Element('xmax')
                xmax1.text = '{}'.format(row[3])
                bnd1.insert(1, xmax1)
                
                ymax1 = Element('ymax')
                ymax1.text = '{}'.format(row[4])
                bnd1.insert(1, ymax1)
                
                doc.write('image{}.xml'.format('{}'.format(7496+int(row[0]))), xml_declaration=True)  
        except ValueError:
            pass
            
            
             
            
```
